```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/enhanced_house_price_dataset.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['City']=le.fit_transform(df['City'])
df['Furnishing']=le.fit_transform(df['Furnishing'])
df['Main Road']=le.fit_transform(df['Main Road'])
df['Guest Room']=le.fit_transform(df['Guest Room'])
df['Basement']=le.fit_transform(df['Basement'])
df['Water Supply']=le.fit_transform(df['Water Supply'])
df['Air Conditioning']=le.fit_transform(df['Air Conditioning'])
df['Preferred Tenant']=le.fit_transform(df['Preferred Tenant'])
```

```python
x=df.drop(columns='Price')
y=df['Price']
```

```python
x
```

```python
x
```

```python
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import  LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```

```python
lr.score(x_test,y_test)
```

 Self Learning

```python
from catboost import CatBoostRegressor
from sklearn.metrics import r2_score
categorical_features_indices = [df.columns.get_loc(col) for col in ['City', 'Furnishing', 'Main Road', 'Guest Room', 'Basement', 'Water Supply', 'Air Conditioning', 'Preferred Tenant']]

catboost_reg = CatBoostRegressor(random_state=42, verbose=0, cat_features=categorical_features_indices) # verbose=0 to suppress output during training
catboost_reg.fit(x_train, y_train)

y_pred_catboost = catboost_reg.predict(x_test)
print(y_pred_catboost)
r2_catboost = r2_score(y_test, y_pred_catboost)
print(f'R-squared score for CatBoost Regressor: {r2_catboost}')
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import r2_score
xgb = XGBRegressor(random_state=42)
xgb.fit(x_train, y_train)
y_pred_xgb = xgb.predict(x_test)
r2_xgb = r2_score(y_test, y_pred_xgb)
print(f'R-squared score for XGBoost: {r2_xgb}')
```
