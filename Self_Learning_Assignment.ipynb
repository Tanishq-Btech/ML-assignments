```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/loan_data_set.csv')
```

```python
df
```

```python
df=df.drop(columns=['Loan_ID'])
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder        # import OneHotEncoder from sklearn
oe=OneHotEncoder()
```

```python
df['Gender']=oe.fit_transform(df[['Gender']]).toarray()        # Use of onehotencoder for nominal data
df['Married']=oe.fit_transform(df[['Married']]).toarray()
df['Education']=oe.fit_transform(df[['Education']]).toarray()
df['Self_Employed']=oe.fit_transform(df[['Self_Employed']]).toarray()
df['Property_Area']=oe.fit_transform(df[['Property_Area']]).toarray()
df['Loan_Status']=oe.fit_transform(df[['Loan_Status']]).toarray()
```

```python
df
```

```python
from sklearn.impute import  SimpleImputer                       # Use of SimpleImputer Library for filling missing values
se=SimpleImputer(missing_values=np.nan,strategy='mean')
df['Dependents'] = df['Dependents'].replace('3+', '3')
df['Dependents'] = pd.to_numeric(df['Dependents'])
se.fit(df[['Married', 'Dependents']])
df[['Married', 'Dependents']]=se.transform(df[['Married', 'Dependents']])
```

```python
df[['LoanAmount','Loan_Amount_Term','Credit_History']]=se.fit_transform(df[['LoanAmount','Loan_Amount_Term','Credit_History']])
```
